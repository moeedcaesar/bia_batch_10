```python
# Create a function that adds an 'age_group' column to df_customers, categorizing ages
# as 'Young' (<30), 'Adult' (30-40), or 'Senior' (>40), and returns the updated DataFrame.
def age_group(df):
    conditions = [
        df['age'] < 30,
        (df['age'] >= 30) & (df['age'] <= 40),
        df['age'] > 40
    ]
    choices = ['Young', 'Adult', 'Senior']
    df['age_group'] = np.select(conditions, choices, default='Unknown')
    return df
```

```python
customers_level = age_group(df_customers)
customers_level
```

```python
# Create a function that normalizes the 'salary' column in df_employees to a 0-1 scale and
# returns the DataFrame with a new 'normalized_salary' column.

def normalize_salary(df):
    maximum = df['salary'].max()
    minimum = df['salary'].min()

    df['normalized'] = (df['salary'] - minimum) / (maximum - minimum)
    return df
```

```python
normalized_salary = normalize_salary(df_employees)
normalized_salary
```

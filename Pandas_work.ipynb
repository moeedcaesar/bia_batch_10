```python
# 2. Difference between merge, join, concat

# Merge is used like 'Joins' of SQL; you can merge two tables based on a related column key
# Join is used to joins tables based on the indexes of two tables irrespective of keys/values in the column
# Concat is used to expand the table horizontally or vertically; it basically increase the length of the table
```

```python
# 3. How to perform a left, right and an inner and outer join between two dataframes

# LEFT join
pd.merge(df1, df2, on='emp_id', how='left')
```

benifits of using categories
    1. Saves Memmory: Text strings are heavy, category identifies each value with a small number under the hood
    2. Faster Processing: Because you work with tiny codes not full texts
    3. Cleaner Data: There are only valid categories allowed. No spelling mistakes, no random new entries
    4. Logical ordering: You can set an order logically like (low < medium < hard) instead of order of alphabets

```python
# 11. How to stack and unstack a dataframe with a multi-index
import numpy as np
index = [('Company A', 2018), ('Company A', 2019),
        ('Company B', 2018), ('Company B', 2019),
        ('Company C', 2018), ('Company C', 2019)]
sales = [125, 211,
         390, 455,
         475, 655]

s = pd.Series(sales, index=index)

m_index = pd.MultiIndex.from_tuples(index)

s = s.reindex(m_index)

s_df = s.unstack()
s_df
```

```python
# 16. How do you use str method to manipulate string data in a dataframe column
df1.astype('string')

# converts to lowercase
lower_case = df1['name'].str.lower()
lower_case

# converts to uppercase
upper_case = df1['name'].str.upper()
upper_case

# to remove the spaces at start and end
space_rmv = df1['name'].str.strip()
space_rmv

# splits the string at given place
df1['Split_Names'] = df1['name'].str.split('a')
df1[['name', 'Split_Names']]

# calculates the length of string
length = df1['name'].str.len()
length

# concatinates the column into a list
concatinated = df1['name'].str.cat(sep=', ')
concatinated

# Gives you boolan to show you if it exists or not in the relative index
index = df1['name'].str.get_dummies()
index

# Returns true if the string starts with given pattern
starts_with = df1['name'].str.startswith('A')
starts_with

# Returns true if the string ends with the give pattern
ends_with = df1['name'].str.endswith('e')
ends_with

# Replace the asked value with given value
replaced = df1['name'].str.replace('Alice', 'Moeed')
replaced

# Repeats the value for a given number
repeated = df1['name'].str.repeat(3)
repeated

# Returns the count of the given element in a string
count = df1['name'].str.count('a')
count

# Gives the position of the first occurance of the pattern
pattern = df1['name'].str.find('a')
pattern

# Gives the list of occurance of patterns
list_of_occurance = df1['name'].str.findall('a')
list_of_occurance

# Checks all the elements in the string are lowercase or not
is_lower = df1['name'].str.islower()
is_lower

# Checks all the elements in the string are uppercase or not
is_upper = df1['name'].str.isupper()
is_upper

# Checks if all the elements in the string are numeric or not
is_number = df1['name'].str.isnumeric()
is_number

# Swaps the lowercase to upper and vise-versa
swap_case = df1['name'].str.swapcase()
swap_case
```

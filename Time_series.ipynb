```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np
```

```python
df = pd.read_csv('yahoo_stock.csv')
```

```python
date = df['Date']
```

```python
closing = df['Close']
```

```python
plt.plot(date, closing)
plt.title('Closing price by date')
step = 365
plt.xticks(date[::step], rotation=30)
plt.xlabel('Date')
plt.ylabel('Closing price')
plt.show
```

```python
df['moving_avg'] = df['Close'].rolling(window=100).mean()

plt.plot(date, closing, label='Closing Price')
plt.plot(date, df['moving_avg'], label='100-day Moving Avg', color='orange')

plt.title('Closing Price with Moving Average')
step = 365
plt.xticks(date[::step], rotation=30, ha='right')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
df['daily_change'] = df['Close'].diff()

# Plot daily change
plt.plot(date, df['daily_change'], label='Daily Change', color='red')

plt.title('Daily Price Change')
step = 100
plt.xticks(date[::step], rotation=30, ha='right')
plt.xlabel('Date')
plt.ylabel('Change in Price')
plt.legend()
plt.show()
```
